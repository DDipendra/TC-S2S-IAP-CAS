```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
from matplotlib.ticker import MultipleLocator, AutoMinorLocator


fig = plt.figure(figsize=(8, 4))

m = Basemap(projection='merc', 
            llcrnrlat=-51, urcrnrlat=51,
            llcrnrlon=0, urcrnrlon=360,
            resolution='l')

polygon_coordsi = np.array([[30, 0], [100, 0], [100, 31], [30, 31], [30, 0]]) 

x_i, y_i = m(polygon_coordsi[:, 0], polygon_coordsi[:, 1])
m.plot(x_i, y_i, marker=None, color='k', linewidth=1)

#reate basin using the polygons points
polygon_coordss = np.array([[100, 0], [180, 0],[180, 40],[180, 40],[100, 40],[100, 0]])
polygon_coords = np.array([[180, 0], [290, 0], [260, 20], [260, 40], [180, 40], [180, 0]])
polygon_coords1 = np.array([[290, 0], [260, 20], [260, 40], [260, 40], [360, 40], [360, 0], [290, 0]])
polygon_coordsh = np.array([[30, 0], [30, -40],[260, -40],[260, 0]])

x2, y2 = m(polygon_coordss[:, 0], polygon_coords[:, 1])
x, y = m(polygon_coords[:, 0], polygon_coords[:, 1])
x1, y1 = m(polygon_coords1[:, 0], polygon_coords1[:, 1])
x4, y4 = m(polygon_coordsh[:, 0], polygon_coordsh[:, 1])

m.plot(x2, y2, marker=None, color='k', linewidth=1)
m.plot(x, y, marker=None, color='k', linewidth=1)
m.plot(x1, y1, marker=None, color='k', linewidth=1)
m.plot(x4, y4, marker=None, color='k', linewidth=1)

lat_ticks = np.arange(-50, 60, 20)
lon_ticks = np.arange(0, 380, 40)

for lat in lat_ticks:
    x, y = m([0, 360], [lat, lat])
    m.plot(x, y, color='black', linewidth=0.001, zorder=10)

for lon in lon_ticks:
    x, y = m([lon, lon], [-50, 60])
    m.plot(x, y, color='black', linewidth=0.001, zorder=10)

for lat in np.arange(-50, 60, 5):
    x, y = m([0, 360], [lat, lat])
    m.plot(x, y, color='black', linewidth=0.001, zorder=10)

for lon in np.arange(0, 380, 5):
    x, y = m([lon, lon], [-50, 60])
    m.plot(x, y, color='black', linewidth=0.001, zorder=10)

m.drawparallels(np.arange(-50, 60, 20), labels=[1,0,0,0], fontsize=10,  color='none') #
m.drawmeridians(np.arange(0, 380, 40), labels=[0,0,0,1], fontsize=10,  color='none')

fig.text(0.22, 0.54, 'NI',fontsize='12',bbox={'facecolor': 'white', 'alpha': 0.7, 'pad': 3})
fig.text(0.4, 0.62, 'NWP',fontsize='12',bbox={'facecolor': 'white', 'alpha': 0.7, 'pad': 3})
fig.text(0.6, 0.6, 'EP',fontsize='12',bbox={'facecolor': 'white', 'alpha': 0.7, 'pad': 3})
fig.text(0.82, 0.6, 'NA',fontsize='12',bbox={'facecolor': 'white', 'alpha': 0.7, 'pad': 3})
fig.text(0.24, 0.35, 'SH',fontsize='12',bbox={'facecolor': 'white', 'alpha': 0.7, 'pad': 3})

m.shadedrelief()
fig.tight_layout()
#plt.savefig('Figures/Study_Area_Basins.png', dpi=400, bbox_inches='tight')
plt.show()
```
