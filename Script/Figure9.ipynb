```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

df = pd.read_csv('Data_TC/Proc/BSS_Basinwise.csv')

fig, axes = plt.subplots(2, 3, figsize=(11, 6), sharex=False)
bar_width = 0.13

df['week_numeric'] = df['week'].str.extract(r'(\d+)').astype(int)

lines = []
lines.append(axes[0, 0].bar(df['week_numeric'] - 2*bar_width, df['WP_m_lag0'], width=bar_width, color='k', label='lag0'))
lines.append(axes[0, 0].bar(df['week_numeric'] - bar_width, df['WP_m_lag0_1'], width=bar_width, color='dodgerblue', label='lag1'))
lines.append(axes[0, 0].bar(df['week_numeric'], df['WP_m_lag0_2'], width=bar_width, color='green', label='lag2'))
lines.append(axes[0, 0].bar(df['week_numeric'] + bar_width, df['WP_m_lag0_3'], width=bar_width, color='red', label='lag3'))
lines.append(axes[0, 0].bar(df['week_numeric'] + 2*bar_width, df['WP_m_lag0_4'], width=bar_width, color='darkred', label='lag4'))
lines.append(axes[0, 0].bar(df['week_numeric'] + 3*bar_width, df['WP_m_lag0_5'], width=bar_width, color='orange', label='lag5'))

axes[0, 0].set_title('NWP', fontsize=12)
axes[0, 0].set_ylim(-0.6, 0.25)
axes[0, 0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 0].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 0].yaxis.set_minor_locator(AutoMinorLocator(6))
axes[0, 0].set_xticks(range(1, 9))
axes[0, 0].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10, rotation=0)
axes[0, 0].tick_params(axis='y', labelsize=10)

# Plot EP
axes[0, 1].bar(df['week_numeric'] - 2*bar_width, df['EP_m_lag0'], width=bar_width, color='k', label='lag0')
axes[0, 1].bar(df['week_numeric'] - bar_width, df['EP_m_lag0_1'], width=bar_width, color='dodgerblue', label='lag1')
axes[0, 1].bar(df['week_numeric'], df['EP_m_lag0_2'], width=bar_width, color='green', label='lag2')
axes[0, 1].bar(df['week_numeric'] + bar_width, df['EP_m_lag0_3'], width=bar_width, color='red', label='lag3')
axes[0, 1].bar(df['week_numeric'] + 2*bar_width, df['EP_m_lag0_4'], width=bar_width, color='darkred', label='lag4')
axes[0, 1].bar(df['week_numeric'] + 3*bar_width, df['EP_m_lag0_5'], width=bar_width, color='orange', label='lag5')

axes[0, 1].set_title('EP', fontsize=12)
axes[0, 1].set_ylim(-0.6, 0.25)
axes[0, 1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 1].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 1].yaxis.set_minor_locator(AutoMinorLocator(6))
axes[0, 1].set_xticks(range(1, 9))
axes[0, 1].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10, rotation=0)
axes[0, 1].tick_params(axis='y', labelsize=10)

# Plot SH
axes[1, 0].bar(df['week_numeric'] - 2*bar_width, df['NI_m_lag0'], width=bar_width, color='k', label='lag0')
axes[1, 0].bar(df['week_numeric'] - bar_width, df['NI_m_lag0_1'], width=bar_width, color='dodgerblue', label='lag1')
axes[1, 0].bar(df['week_numeric'], df['NI_m_lag0_2'], width=bar_width, color='green', label='lag2')
axes[1, 0].bar(df['week_numeric'] + bar_width, df['NI_m_lag0_3'], width=bar_width, color='red', label='lag3')
axes[1, 0].bar(df['week_numeric'] + 2*bar_width, df['NI_m_lag0_4'], width=bar_width, color='darkred', label='lag4')
axes[1, 0].bar(df['week_numeric'] + 3*bar_width, df['NI_m_lag0_5'], width=bar_width, color='orange', label='lag5')

axes[1, 0].set_title('NI', fontsize=12)
axes[1, 0].set_ylim(-0.6, 0.25)
axes[1, 0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1, 0].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[1, 0].yaxis.set_minor_locator(AutoMinorLocator(6))
axes[1, 0].set_xticks(range(1, 9))
axes[1, 0].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10, rotation=0)
axes[1, 0].tick_params(axis='y', labelsize=10)

# Plot NA
axes[0, 2].bar(df['week_numeric'] - 2*bar_width, df['NA_m_lag0'], width=bar_width, color='k', label='lag0')
axes[0, 2].bar(df['week_numeric'] - bar_width, df['NA_m_lag0_1'], width=bar_width, color='dodgerblue', label='lag1')
axes[0, 2].bar(df['week_numeric'], df['NA_m_lag0_2'], width=bar_width, color='green', label='lag2')
axes[0, 2].bar(df['week_numeric'] + bar_width, df['NA_m_lag0_3'], width=bar_width, color='red', label='lag3')
axes[0, 2].bar(df['week_numeric'] + 2*bar_width, df['NA_m_lag0_4'], width=bar_width, color='darkred', label='lag4')
axes[0, 2].bar(df['week_numeric'] + 3*bar_width, df['NA_m_lag0_5'], width=bar_width, color='orange', label='lag5')

axes[0, 2].set_title('NA', fontsize=12)
axes[0, 2].set_ylim(-0.6, 0.25)
axes[0, 2].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 2].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 2].yaxis.set_minor_locator(AutoMinorLocator(6))
axes[0, 2].set_xticks(range(1, 9))
axes[0, 2].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10, rotation=0)
axes[0, 2].tick_params(axis='y', labelsize=10)

# Plot SH
axes[1, 1].bar(df['week_numeric'] - 2*bar_width, df['SH_m_lag0'], width=bar_width, color='k', label='lag0')
axes[1, 1].bar(df['week_numeric'] - bar_width, df['SH_m_lag0_1'], width=bar_width, color='dodgerblue', label='lag1')
axes[1, 1].bar(df['week_numeric'], df['SH_m_lag0_2'], width=bar_width, color='green', label='lag2')
axes[1, 1].bar(df['week_numeric'] + bar_width, df['SH_m_lag0_3'], width=bar_width, color='red', label='lag3')
axes[1, 1].bar(df['week_numeric'] + 2*bar_width, df['SH_m_lag0_4'], width=bar_width, color='darkred', label='lag4')
axes[1, 1].bar(df['week_numeric'] + 3*bar_width, df['SH_m_lag0_5'], width=bar_width, color='orange', label='lag5')

axes[1, 1].set_title('SH', fontsize=12)
axes[1, 1].set_ylim(-0.6, 0.25)
axes[1, 1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1, 1].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[1, 1].yaxis.set_minor_locator(AutoMinorLocator(6))
axes[1, 1].set_xticks(range(1, 9))
axes[1, 1].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10, rotation=0)
axes[1, 1].tick_params(axis='y', labelsize=10)

# Remove the empty subplot
fig.delaxes(axes[1, 2])

for ax in axes.flat:
    for _, spine in ax.spines.items():
        spine.set_linewidth(1)

axes[0, 0].axhline(0, color='k', alpha=1, linewidth=0.6, linestyle='dotted')
axes[0, 1].axhline(0, color='k', alpha=1, linewidth=0.6, linestyle='dotted')
axes[0, 2].axhline(0, color='k', alpha=1, linewidth=0.6, linestyle='dotted')
axes[1, 0].axhline(0, color='k', alpha=1, linewidth=0.6, linestyle='dotted')
axes[1, 1].axhline(0, color='k', alpha=1, linewidth=0.6, linestyle='dotted')

axes[0, 0].text(0.4, 0.18, '(a)', fontsize=12)
axes[0, 1].text(0.4, 0.18, '(b)', fontsize=12)
axes[0, 2].text(0.4, 0.18, '(c)', fontsize=12)
axes[1, 0].text(0.4, 0.18, '(d)', fontsize=12)
axes[1, 1].text(0.4, 0.18, '(e)', fontsize=12)


legend_axes = fig.add_subplot(2, 3, 6)  
legend_axes.axis('off')  
legend_axes.legend(lines, ['lag0', 'lag1', 'lag2', 'lag3', 'lag4', 'lag5'], loc='upper left', 
                    ncol=2, facecolor='white', fontsize=12)

plt.tight_layout()
#plt.savefig('Figures/Figure9_lag_ensembles.png', dpi=400, bbox_inches='tight')
plt.show()
```
