```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

df = pd.read_csv('Data_TC/Proc/BSS_Basinwise.csv')

fig, axes = plt.subplots(2, 3, figsize=(10, 5), sharex=False)

# Plot WP
lines = [] 

lines.append(axes[0, 0].plot(df['week'], df['forlegend'], linestyle='-', color='k',linewidth=1.5,label='lag0')[0]) 
lines.append(axes[0, 0].plot(df['week'], df['forlegend'], linestyle='-', color='dodgerblue',linewidth=1.5,label='lag4')[0]) 
lines.append(axes[0, 0].plot(df['week'], df['forlegend'], linestyle='-', color='red',linewidth=1.5,label='lag1')[0]) 
lines.append(axes[0, 0].plot(df['week'], df['forlegend'], linestyle='-', color='green',linewidth=1.5,label='lag4')[0]) 

lines.append(axes[0, 0].plot(df['week'], df['WP_m_lag0'], linestyle='-', color='k', linewidth=1, label='lag0')[0])
lines.append(axes[0, 0].plot(df['week'], df['WP_m_lag0_cal'], linestyle='-', color='dodgerblue', linewidth=1)[0])
lines.append(axes[0, 0].plot(df['week'], df['WP_m_lag0_cal_lin'], linestyle='-', color='red', linewidth=1)[0])
lines.append(axes[0, 0].plot(df['week'], df['WP_m_lag0_cal_lin_out'], linestyle='-', color='green', linewidth=1)[0])

axes[0, 0].set_title('NWP', fontsize=12)
axes[0, 0].set_ylim(-0.7, 0.5)
axes[0, 0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 0].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 0].yaxis.set_minor_locator(AutoMinorLocator())
axes[0, 0].set_xticks(range(0, 8))
axes[0, 0].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10)
axes[0, 0].tick_params(axis='y', labelsize=10)

# Plot EP
axes[0, 1].plot(df['week'], df['EP_m_lag0'], linestyle='-', color='k', linewidth=1)
axes[0, 1].plot(df['week'], df['EP_m_lag0_cal'], linestyle='-', color='dodgerblue', linewidth=1)
axes[0, 1].plot(df['week'], df['EP_m_lag0_cal_lin'], linestyle='-', color='red', linewidth=1)
axes[0, 1].plot(df['week'], df['EP_m_lag0_cal_lin_out'], linestyle='-', color='green', linewidth=1)

axes[0, 1].set_title('EP', fontsize=12)
axes[0, 1].set_ylim(-0.7, 0.5)
axes[0, 1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 1].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 1].yaxis.set_minor_locator(AutoMinorLocator())
axes[0, 1].set_xticks(range(0, 8))
axes[0, 1].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10)
axes[0, 1].tick_params(axis='y', labelsize=10)

# Plot SH
axes[1, 0].plot(df['week'], df['NI_m_lag0'], linestyle='-', color='k', linewidth=1)
axes[1, 0].plot(df['week'], df['NI_m_lag0_cal'], linestyle='-', color='dodgerblue', linewidth=1)
axes[1, 0].plot(df['week'], df['NI_m_lag0_cal_lin'], linestyle='-', color='red', linewidth=1)
axes[1, 0].plot(df['week'], df['NI_m_lag0_cal_lin_out'], linestyle='-', color='green', linewidth=1)

axes[1, 0].set_title('NI', fontsize=12)
axes[1, 0].set_ylim(-0.7, 0.5)
axes[1, 0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1, 0].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[1, 0].yaxis.set_minor_locator(AutoMinorLocator())
axes[1, 0].set_xticks(range(0, 8))
axes[1, 0].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10)
axes[1, 0].tick_params(axis='y', labelsize=10)

# Plot NA
axes[0, 2].plot(df['week'], df['NA_m_lag0'], linestyle='-', color='k', linewidth=1)

axes[0, 2].plot(df['week'], df['NA_m_lag0_cal'], linestyle='-', color='dodgerblue', linewidth=1)
axes[0, 2].plot(df['week'], df['NA_m_lag0_cal_lin'], linestyle='-', color='red', linewidth=1)
axes[0, 2].plot(df['week'], df['NA_m_lag0_cal_lin_out'], linestyle='-', color='green', linewidth=1)

axes[0, 2].set_title('NA', fontsize=12)
axes[0, 2].set_ylim(-0.7, 0.5)
axes[0, 2].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0, 2].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[0, 2].yaxis.set_minor_locator(AutoMinorLocator())
axes[0, 2].set_xticks(range(0, 8))
axes[0, 2].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10)
axes[0, 2].tick_params(axis='y', labelsize=10)


new_subplot = axes[1, 1]

# Plot EP
axes[1, 1].plot(df['week'], df['SH_m_lag0'], linestyle='-', color='k', linewidth=1)
axes[1, 1].plot(df['week'], df['SH_m_lag0_cal'], linestyle='-', color='dodgerblue', linewidth=1)
axes[1, 1].plot(df['week'], df['SH_m_lag0_cal_lin'], linestyle='-', color='red', linewidth=1)
axes[1, 1].plot(df['week'], df['SH_m_lag0_cal_lin_out'], linestyle='-', color='green', linewidth=1)

axes[1, 1].set_title('SH', fontsize=12)
axes[1, 1].set_ylim(-0.7, 0.5)
axes[1, 1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1, 1].xaxis.set_minor_locator(plt.MultipleLocator(1))
axes[1, 1].yaxis.set_minor_locator(AutoMinorLocator())
axes[1, 1].set_xticks(range(0, 8))
axes[1, 1].set_xticklabels(['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8'], fontsize=10)
axes[1, 1].tick_params(axis='y', labelsize=10)

# Remove the empty subplot
fig.delaxes(axes[1, 2])

for ax in axes.flat:
    for _, spine in ax.spines.items():
        spine.set_linewidth(1)

axes[0, 0].axhline(0, color='gray', alpha=1, linewidth=0.6, linestyle='-')
axes[0, 1].axhline(0, color='gray', alpha=1, linewidth=0.6, linestyle='-')
axes[1, 1].axhline(0, color='gray', alpha=1, linewidth=0.6, linestyle='-')
axes[1, 0].axhline(0, color='gray', alpha=1, linewidth=0.6, linestyle='-')
axes[0, 2].axhline(0, color='gray', alpha=1, linewidth=0.6, linestyle='-')

axes[0, 0].text(-0.2, 0.38, '(a)', fontsize=12,)
axes[0, 1].text(-0.2, 0.38, '(b)', fontsize=12,)
axes[1, 0].text(-0.2, 0.38, '(d)', fontsize=12,)
axes[1, 1].text(-0.2, 0.38, '(e)', fontsize=12,)
axes[0, 2].text(-0.2, 0.38, '(c)', fontsize=12,)



legend_axes = fig.add_subplot(2, 3, 6) 
legend_axes.axis('off')  

legend_axes.legend(lines, ['BSSm','BSSm|rmean', 'BSSm|linear','BSSm|linear_out'], loc='upper left', ncols=1, facecolor='white',fontsize=12)

plt.tight_layout()
#plt.savefig('Figures/Figure7_Calibration.png', dpi=400, bbox_inches='tight')

plt.show()
```
