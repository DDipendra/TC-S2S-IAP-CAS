```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Data_TC/Proc/Annual_climatology.csv')
data.set_index('Basin', inplace=True)

colors = ['dimgray', 'orangered', 'dodgerblue', 'darkgoldenrod', 'seagreen', 'yellowgreen', 'royalblue', 'Purple','darkorange']

fig, ax = plt.subplots(figsize=(6, 3.5))
data.plot(kind='bar', ax=ax, color=colors, width=0.8)

plt.ylabel('TC Number')

legend_labels = data.columns.tolist()
legend_patches = [plt.Rectangle((0,0),1,1, color=color) for color in colors]
plt.legend(legend_patches, legend_labels, loc='upper left', bbox_to_anchor=(0.07, 0.98), ncol=3,fontsize='10')

ax.set_ylim(0, 90) 

for spine in ax.spines.values():
    spine.set_linewidth(1)

ax.set_xticklabels(data.index, rotation=0, ha='center')    
    
# Show the plot
plt.tight_layout()
#plt.savefig('Figures/Figure2_climatology.png', dpi=400, bbox_inches='tight')
plt.show()
```
